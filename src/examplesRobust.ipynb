##Â Complete Graphs

```python
# Create a complete graph
n = 10
G = create_complete(n)

# Call the algorithm to simplify the graph
qr = 3
G2 = simplify_graph(G, qr, displays=True)

# Draw the graphs and plot
title = f'Complete graph of size {n} with quantifier rank {qr}'
draw_graphs(G, G2, title)
```

##Â Trees

```python
# Create a balanced tree
h = 3
G = create_tree(h=h, children=[2,3,4])

# Call the algorithm to simplify the graph
qr = 2
G2 = simplify_graph(G, qr, displays=True)

# Draw the graphs and plot
title = f'Tree of height {h} with quantifier rank {qr}'
draw_graphs(G, G2, title)
```

## Random Graphs

```python
# Create a random graph
n = 10       # Graphs size
p = 0.7      # Probablility of an edge
G = create_random(n, p)

# Call the algorithm to simplify the graph
qr = 1
G2 = simplify_graph(G, qr, displays=True)

# Draw the graphs and plot
title = f'Random graph of size n={n} and probability of having an edge p={p}'
draw_graphs(G, G2, title)
```
